## 5478. 最大得分
---
链接:[https://leetcode-cn.com/contest/weekly-contest-200/problems/get-the-maximum-score/](https://leetcode-cn.com/contest/weekly-contest-200/problems/get-the-maximum-score/)

```
两个有序且数组内元素互不相同的数组 nums1 和 nums2 。

一条 合法路径 定义如下：
1. 选择nums1 或 nums2 开始遍历（从下标 0 处开始）。从左到右遍历当前数组。
2. 如果nums1nums2 中都存在值，可以切换路径到另一个数组对应数字处继续遍历（重复数字只会被统计一次）。
3. 得分定义为合法路径中不同数字的和。返回最大得分。

由于答案可能很大，请你将它对 10^9 + 7 取余后返回。
```
example:

![photo](https://assets.leetcode-cn.com/aliyun-lc-upload/uploads/2020/08/02/sample_1_1893.png)
```
输入：nums1 = [2,4,5,8,10], nums2 = [4,6,8,9]
输出：30
解释：合法路径包括：
[2,4,5,8,10], [2,4,5,8,9], [2,4,6,8,9], [2,4,6,8,10],（从 nums1 开始遍历）
[4,6,8,9], [4,5,8,10], [4,5,8,9], [4,6,8,10]  （从 nums2 开始遍历）
最大得分为上图中的绿色路径 [2,4,6,8,10] 。
```
### 代码
```python
class Solution:
    def maxSum(self, nums1: List[int], nums2: List[int]) -> int:
        t = []
        for i in nums1:
            t.append((i,0))
        for i in nums2:
            t.append((i,1))
        t = sorted(t)
        print(t)
        n = len(t)
        f = [0]*n
        i = 0
        fp = [0,0] #fp[0]存储遍历nums1最大值，fp[1]存储遍历nums2最大值
        while i < n:
            if i+1 < n and t[i][0] == t[i+1][0]:
                fp = [max(fp) + t[i][0]]*2 #使fp[0]=fp[1]=max(fp)
                i += 2
            else:
                fp[t[i][1]] += t[i][0] #根据t[][1]的值是0 or 1，存放入对应fp中，为了比较两条路径的最大值。
                i += 1
        return max(fp) % int(1e9 + 7)
```

## 35483. 整理字符串
---
链接：https://leetcode-cn.com/problems/make-the-string-great/

```
字符串s由大小写字母组成，如果想同字符大小写在一起则删除这两个字母。输出处理后的结果。

输入：s = "leEeetcode"
输出："leetcode"
解释：无论你第一次选的是 i = 1 还是 i = 2，都会使 "leEeetcode" 缩减为 "leetcode" 。
```
时间复杂度：O(n),空间复杂度：O(n)

### 代码

```python
class Solution:
    def makeGood(self, s: str) -> str:
        n = len(s)
        ans = [] #结果字符串
        for i in range(n):
            if not ans: #ans为空时，入栈
                ans.append(s[i])
            else: #不为空时，利用ord判断s[i]&ans[-1]是否为大小写
                if abs(ord(s[i]) - ord(ans[-1])) == 32: #ord函数将字母转换成ASCII码，大小写之间相隔32
                    ans.pop()
                else:
                    ans.append(s[i])
        return("".join(ans))
```

## 5484. 找出第 N 个二进制字符串中的第 K 位
---
链接：https://leetcode-cn.com/problems/find-kth-bit-in-nth-binary-string/

```
S1 = "0"
当 i > 1 时，Si = Si-1 + "1" + reverse(invert(Si-1))
```

* reverse(x)反转 x 后得到的字符串，
* invert(x) 则会翻转x中的每一位: 0变为1，而1变为0

* 求：Sn的第k位（k从1开始）

example:
```
S1 = "0"
S2 = "011"
S3 = "0111001"
S4 = "011100110110001"

输入：n = 3, k = 1
输出："0"
```
* 时间复杂度：O(n\*(2^n-1)) ，Sn的长度是2^n-1
* 空间复杂度：O(2^n -1) ,需要Sn长的空间

### 代码
```python
class Solution:
    def findKthBit(self, n: int, k: int) -> str:
        def reverse(S:list):
            #invert翻转
            for i in range(len(S)):
                if S[i] == 0:
                    S[i] = 1
                else:
                    S[i] = 0
            # 反转
            return(S[::-1])
        
        i = 1
        S = [0]
        while i < n:
            S = S + [1] +  reverse(S)
            i += 1
        return(str(S[k-1]))
```

## 5471. 和为目标值的最大数目不重叠非空子数组数目
----
链接：https://leetcode-cn.com/problems/maximum-number-of-non-overlapping-subarrays-with-sum-equals-target/

给你一个数组 nums 和一个整数 target 。

请你返回 非空不重叠 子数组的最大数目，且每个子数组中数字和都为 target 。

example:
```
输入：nums = [-1,3,5,1,4,2,-9], target = 6
输出：2
解释：总共有 3 个子数组和为 6 。
([5,1], [4,2], [3,5,1,4,2,-9]) 但只有前 2 个是不重叠的。
```

### 代码
```python
class Solution:
    def maxNonOverlapping(self, nums: List[int], target: int) -> int:
        ans = 0
        tmp_sum = set() #存储每个nums[i:j]的结果
        tmp_sum.add(0)
        cur_sum = 0 #当前nums[i:j]的结果
        for num in nums:
            cur_sum += num
            if cur_sum - target in tmp_sum:
                tmp_sum = set()
                tmp_sum.add(0)
                ans += 1
                cur_sum = 0
            else:
                tmp_sum.add(cur_sum)
        return(ans)
```
